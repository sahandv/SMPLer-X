### Match segments and find joints

```python
import numpy as np

def connect_segments_with_tolerance(prox_positions, dist_positions, tolerance=1e-3):
    """
    Connect segments into a chain of joints, handling close but not matching points.

    Parameters:
        prox_positions (array): Proximal end positions of segments.
        dist_positions (array): Distal end positions of segments.
        tolerance (float): Threshold to consider positions as "close enough."

    Returns:
        joints (array): Ordered list of joint positions.
    """
    joints = [prox_positions[0]]  # Start with the first proximal position

    for i in range(len(prox_positions)):
        # Add the distal position of the current segment
        joints.append(dist_positions[i])
        
        # Check if the current distal end matches the next proximal start
        if i + 1 < len(prox_positions):
            if not np.allclose(dist_positions[i], prox_positions[i + 1], atol=tolerance):
                # If not close enough, snap to the midpoint
                midpoint = (dist_positions[i] + prox_positions[i + 1]) / 2.0
                print(f"Gap detected between segment {i} and {i + 1}. Snapping to midpoint: {midpoint}")
                joints[-1] = midpoint  # Replace current distal with the midpoint
                joints.append(midpoint)  # Add midpoint before the next segment starts

    return np.array(joints)

# # Example usage:
# prox_positions = np.array([[0, 0, 0], [5, 0, 0], [8.1, 3, 0]])  # Proximal positions
# dist_positions = np.array([[5, 0, 0], [8, 3, 0], [10, 5, 2]])  # Distal positions

# joints = connect_segments_with_tolerance(prox_positions, dist_positions, tolerance=0.2)
# print("Joint Positions:\n", joints)
```

```python

import pandas as pd
# import matplotlib.pyplot as plt
import plotly.graph_objects as go

distal_data = subject_distal_df.loc[[0,2,7,frame_no]].copy()
proximal_data = subject_proximal_df.loc[[0,2,7,frame_no]].copy()

# drop columns where there is nan
distal_data.dropna(axis=1,inplace=True)
proximal_data.dropna(axis=1,inplace=True)

point_names_proximal = proximal_data.loc[0, [i for i, x in (proximal_data.loc[7] == "X").items() if x]].values.tolist()#2
point_names_distal = distal_data.loc[0, [i for i, x in (distal_data.loc[7] == "X").items() if x]].values.tolist()#2

# Prepare the data (using your existing data extraction logic)
xp = list(map(float, proximal_data[[i for i, x in (proximal_data.loc[7] == "X").items() if x]].loc[frame_no].values.tolist()))
xd = list(map(float, distal_data[[i for i, x in (distal_data.loc[7] == "X").items() if x]].loc[frame_no].values.tolist()))

yp = list(map(float, proximal_data[[i for i, x in (proximal_data.loc[7] == "Y").items() if x]].loc[frame_no].values.tolist()))
yd = list(map(float, distal_data[[i for i, x in (distal_data.loc[7] == "Y").items() if x]].loc[frame_no].values.tolist()))

zp = list(map(float, proximal_data[[i for i, x in (proximal_data.loc[7] == "Z").items() if x]].loc[frame_no].values.tolist()))
zd = list(map(float, distal_data[[i for i, x in (distal_data.loc[7] == "Z").items() if x]].loc[frame_no].values.tolist()))


# Find the min and max values across all axes to set consistent ranges
x_min = min(min(xp), min(xd))
x_max = max(max(xp), max(xd))
y_min = min(min(yp), min(yd))
y_max = max(max(yp), max(yd))
z_min = min(min(zp), min(zd))
z_max = max(max(zp), max(zd))

# Determine a common range for all axes
axis_range = [
    min(x_min, y_min, z_min),  # Minimum range value across all axes
    max(x_max, y_max, z_max)  # Maximum range value across all axes
]

# Create the Plotly 3D scatter plot
fig = go.Figure()

# Add distal points (blue)
fig.add_trace(go.Scatter3d(
    x=xd, y=yd, z=zd,
    mode='markers+text',
    marker=dict(size=5, color='blue'),
    text=point_names_distal,  # Add point names as hover text
    textposition="top center",
    name='Distal'
))

# Add proximal points (red)
fig.add_trace(go.Scatter3d(
    x=xp, y=yp, z=zp,
    mode='markers+text',
    marker=dict(size=5, color='red'),
    text=point_names_proximal,  # Add point names as hover text
    textposition="top center",
    name='Proximal'
))

# Draw edges (lines) between proximal and distal points with the same name
for i, name in enumerate(point_names_proximal):
    if name in point_names_distal:  # Check if the name exists in the distal labels
        j = point_names_distal.index(name)  # Find the corresponding index
        fig.add_trace(go.Scatter3d(
            x=[xp[i], xd[j]],
            y=[yp[i], yd[j]],
            z=[zp[i], zd[j]],
            mode='lines',
            line=dict(color='gray', width=2),
            showlegend=False  # Hide legend for individual lines
        ))


# Update layout for better visualization
fig.update_layout(
    scene=dict(
        xaxis=dict(title='X Axis', range=axis_range),
        yaxis=dict(title='Y Axis', range=axis_range),
        zaxis=dict(title='Z Axis', range=axis_range),
        aspectmode='cube'  # Ensures a cubic aspect ratio
    ),
    title='Interactive 3D Points with Consistent Axis Scaling',
    height=800  # Adjust height to make the plot taller
)

# Show the interactive plot
fig.show()
```
